```python
import pandas as pd
import pydicom
import matplotlib.pyplot as plt
import random
```

```python
# 读取
train_series_descriptions_df = pd.read_csv('./dataset/train_series_descriptions.csv')

# 显示
train_series_descriptions_df.head()
# 6294 rows × 3 columns
```

```python
# 查看 series_description 列有几种不同的值
unique_descriptions = train_series_descriptions_df['series_description'].unique()
description_counts = train_series_descriptions_df['series_description'].value_counts()

# 输出不同的 series_description 值及其数量
unique_descriptions, description_counts

# 作出直方分布图
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.countplot(data=train_series_descriptions_df, y='series_description', order=description_counts.index)
plt.title('Distribution of Series Descriptions')
plt.xlabel('Count')
plt.ylabel('Series Description')
plt.show()
```
